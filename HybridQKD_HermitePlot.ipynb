```python
from scipy import special

import matplotlib.pyplot as plt

import numpy as np
```

```python
#Single photon case
p_monic = special.hermite(1)
q_monic = special.hermite(0)

x = np.linspace(-3, 3, 400)
x1= np.linspace(-3, 3, 400)
y = (np.exp(-(x**2)/2)*p_monic(x/np.sqrt(2))**2)/(2*np.sqrt(2))
y1= (np.exp(-(x1**2)/2)*q_monic(x1/np.sqrt(2))**2)/(np.sqrt(2))
plt.plot(x, y,x1,y1)

plt.title(" Hermite polynomial")

plt.xlabel("x")

plt.ylabel("y")

plt.show()
```

```python
# Coherent state
c = np.linspace(3, 4, 400)
c1= np.linspace(3, 4, 400)
d = np.exp(-(c**2)/2)/(np.sqrt(2)*np.pi)
d1= (np.exp(-(c1-(2*0.34*1/np.sqrt(2)))**2/2))/(np.sqrt(2))
d2= (np.exp(-(c1-(2*0.44*1/np.sqrt(2)))**2/2))/(np.sqrt(2))
plt.plot(c,d,c1, d1,c1,d2)


plt.show()

```
