```python
#with IR
def skr(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-e
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps
# with MED    
def skr1(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-3*e
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps
# with Cloning
def skr2(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-5*e 
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps
        
    
def skr3(l):
    mu=0.2
    gamma= 10**7
    s=(2)/(16*3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)+0.029289
    tau=1-((3/4)*mu*(1-t(l)))-e
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps
        
    
def skr4(l):
    mu=0.03
    gamma= 10**7
    s=(2/3)/16
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)+0.012
    tau=1-((3/4)*mu*(1-t(l)))-e
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps

def hfunc(e):
    h=-e*np.log2(e)-((1-e)*np.log2(1-e))
    return h

def skrup(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-e
    Rdps=s*gamma*pclick*(1-hfunc(e)-hfunc((3+np.sqrt(5))*e))
    return Rdps

    
```

```python
import math
def skrlow(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    k=1 - e**2 - (1 - (6*(e**2)))/2
    tau=10*e*np.log2(k)
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps

```

```python
def qber(mu):
    psuc=float(1-np.exp(-2*mu))
    qber=float((psuc/2)+(2/3)*(1-psuc)**2)
    return qber
def evebits(mu):
    psuc=1-np.exp(-2*mu)
    bitsknown=(2*psuc)/3
    evebits = bitsknown*(1/qber(mu))
    return evebits
def skrwcs(mu,l):
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-(evebits(mu)*e)
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps
    
def skrwcsir(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-(e*mu)
    Rdps=s*gamma*pclick*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps

def skr1photon(l):
    mu=0.2
    gamma= 10**7
    s=(2/3)
    a=0.1
    b=0.01
    ls=1
    eta=0.1
    def t(l):
        ls=1
        eta=0.1
        t=eta*10**((-1*((0.2*l)+ls))/10)
        return t
    fe=1.16
    d=1.95*(10**(-5))
    psignal=mu*t(l)
    pdark=2*d
    pclick=psignal+ pdark
    e=((0.5*pdark+b*psignal)/psignal)
    tau=1-((3/4)*mu*(1-t(l)))-(e)
    Rdps=s*gamma*(tau+fe*((e*np.log2(e))+((1-e)*np.log2(1-e))))
    return Rdps
```
